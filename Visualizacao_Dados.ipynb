```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.iris()
df
```

```python
fig = px.bar(df, x="sepal_width", y="sepal_length")
fig.show()
```

```python
fig = px.bar(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
fig = px.bar(df, x="sepal_width", y="sepal_length", color="species", barmode="group")
fig.show()
```

```python
fig = px.bar(df, x="sepal_width", y="sepal_length", color="species", barmode="overlay")
fig.show()
```

```python
fig = px.bar(df, x="sepal_width", y="sepal_length", color="species", barmode="group", facet_row="species", facet_col="species_id")
fig.show()
```
